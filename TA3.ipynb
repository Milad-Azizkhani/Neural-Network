```python
x = np.linspace(0,30,100)
y = np.sin(x)
z = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x, y, label='sin')
plt.legend(loc='best')

plt.subplot(1,2,2)
plt.plot(x, z, label='cos')

plt.legend(loc='best')
plt.grid()
```
